```python
import numpy as np
import matplotlib.pyplot as plt
import time # ilem sresi iin

# Rastgele Nokta retimi Yapalm

start = time.time()
n = 10000  # Nokta Says
x = np.random.uniform(-1, 1, n)  # -1, 1 aralnda n adet restgele x ve y kordinat oluturur
y = np.random.uniform(-1, 1, n)

# ember indeki Noktalarn Hesaplanmas

distance = np.sqrt(x**2 + y**2)  # her bir noktann orijine uzakl
inside_circle = distance <= 1  # ember iinde olan noktalar (mesafe <= 1)

# Pi saysnn hesaplanmas

pi = 4 * np.sum(inside_circle) / n  # emberin iinde ka nokta olduunu sayar. Pi'nin yaklak deerini verir.
print(f"Monte Carlo ile Hesaplanan Pi: {pi}")

# Sonularn Grselletirilmesi

plt.figure(figsize = (6, 6))
plt.scatter(x[inside_circle], y[inside_circle], color="blue", s = 1, label="ember i") # emberin iindeki noktalar mavi
plt.scatter(x[~inside_circle], y[~inside_circle], color="red", s = 1, label="ember D") # emberin dndaki noktalar krmz
plt.axhline(0, color="black", linewidth=0.5) # X Ekseni izer
plt.axvline(0, color="black", linewidth=0.5) # Y Ekseni izer
plt.legend()
plt.title("Monte Carlo ile Pi Hesaplama")
plt.savefig("circle_simulation.png") # Grafii kaydeder
plt.show()
end = time.time()
print(f"lem Sresi: {end - start} saniye")
```
