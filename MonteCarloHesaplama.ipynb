```python
import numpy as np
import matplotlib.pyplot as plt
import time # işlem süresi için

# Rastgele Nokta Üretimi Yapalım

start = time.time()
n = 10000  # Nokta Sayısı
x = np.random.uniform(-1, 1, n)  # -1, 1 aralığında n adet restgele x ve y kordinatı oluşturur
y = np.random.uniform(-1, 1, n)

# Çember İçindeki Noktaların Hesaplanması

distance = np.sqrt(x**2 + y**2)  # her bir noktanın orijine uzaklığı
inside_circle = distance <= 1  # çember içinde olan noktalar (mesafe <= 1)

# Pi sayısının hesaplanması

pi = 4 * np.sum(inside_circle) / n  # Çemberin içinde kaç nokta olduğunu sayar. Pi'nin yaklaşık değerini verir.
print(f"Monte Carlo ile Hesaplanan Pi: {pi}")

# Sonuçların Görselleştirilmesi

plt.figure(figsize = (6, 6))
plt.scatter(x[inside_circle], y[inside_circle], color="blue", s = 1, label="Çember İçi") # Çemberin içindeki noktalar mavi
plt.scatter(x[~inside_circle], y[~inside_circle], color="red", s = 1, label="Çember Dışı") # Çemberin dışındaki noktalar kırmızı
plt.axhline(0, color="black", linewidth=0.5) # X Ekseni Çizer
plt.axvline(0, color="black", linewidth=0.5) # Y Ekseni Çizer
plt.legend()
plt.title("Monte Carlo ile Pi Hesaplama")
plt.savefig("circle_simulation.png") # Grafiği kaydeder
plt.show()
end = time.time()
print(f"İşlem Süresi: {end - start} saniye")
```
